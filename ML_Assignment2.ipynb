```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV,StratifiedKFold
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
vehicle=pd.read_csv("Vehicle.csv")
vehicle.head()
```

```python
vehicle.Class.unique()
```

```python
lbl=LabelEncoder()
vehicle.Class=lbl.fit_transform(vehicle.Class)
vehicle.head()
```

```python
X=vehicle.drop('Class',axis=1)
y=vehicle['Class']
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

# 1. SVM w/o scaling Linear

```python
svm=SVC(kernel='linear',probability=True,random_state=23,)
```

```python
params={'C':np.linspace(0.001,5,10)}
```

```python
gcv=GridSearchCV(svm,param_grid=params,cv=kfold,verbose=3,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
print("Best Parameters: ",gcv.best_params_)
print("Best Score: ",gcv.best_score_)
```

# 2.SVM w/o scaling RBF

```python
svm=SVC(kernel='rbf',probability=True,random_state=23,)
```

```python
params={'C':np.linspace(0.001,5,30),'gamma':list(np.linspace(0.001,5,10))+['scale','auto']}
```

```python
gcv=GridSearchCV(svm,param_grid=params,cv=kfold,verbose=3,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print("Best Parameters: ",gcv.best_params_)
print("Best Score: ",gcv.best_score_)
```

# 3. SVM with scaling Linear

```python
scaler=StandardScaler()
```

```python
svm_scl_linear=SVC(kernel='linear',probability=True, random_state=23)
```

```python
pipe=Pipeline([('SCL',scaler),('SVM',svm_scl_linear)])
```

```python
pipe.get_params()
```

```python
params={'SCL':[StandardScaler(),MinMaxScaler()],'SVM__C':np.linspace(0.001,5,10)}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,verbose=3,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print("Best Parameters: ",gcv.best_params_)
print("Best Score: ",gcv.best_score_)
```

```python
# As per my observation after scaling SVM-linear gives good result
```

# 4. SVM with scaling RBF

```python
scaler=StandardScaler()
```

```python
svm_scl_rbf=SVC(kernel='rbf',probability=True,random_state=23)
```

```python
pipe=Pipeline([('SCL',scaler),('SVM',svm_scl_rbf)])
```

```python
params={'SCL':[StandardScaler(),MinMaxScaler()],'SVM__C':np.linspace(0.001,5,30),
        'SVM__gamma':list(np.linspace(0.001,5,10))+['scale','auto']}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss',verbose=3)
gcv.fit(X,y)
```

```python
print("Best Parameters: ",gcv.best_params_)
print("Best Score: ",gcv.best_score_)
```

```python
# As per my observation after scaling SVM-rbf gives good result
```
