```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import log_loss,accuracy_score
from sklearn.preprocessing import LabelEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV,StratifiedKFold,KFold
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
vehicle = pd.read_csv('Vehicle.csv')
```

```python
vehicle.head()
```

```python
lbl=LabelEncoder()
vehicle['Class']=lbl.fit_transform(vehicle['Class'])
X = vehicle.drop("Class",axis =1)
y = vehicle['Class']
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
scaler=StandardScaler()
knn=KNeighborsClassifier()
pipe=Pipeline([('SCL',scaler),('KNN',knn)])
```

```python
params={'KNN__n_neighbors':np.arange(1,11)}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

```python
lr = LogisticRegression()
```

```python
pipe=Pipeline([('SCL',scaler),('LR',lr)])
```

```python
pipe.get_params()
```

```python
params={'LR__multi_class':['multinomial','OVR'],'LR__penalty':['l1', 'l2', 'elasticnet', None],
         'LR__solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```
