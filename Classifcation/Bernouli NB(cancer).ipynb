```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,log_loss
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV,StratifiedKFold 
import warnings as ws
ws.filterwarnings("ignore")
```

```python
cancer=pd.read_csv("Cancer.csv",index_col=0)
dum_cancer=pd.get_dummies(cancer,drop_first=True)
dum_cancer.head()
```

```python
X=dum_cancer.drop('Class_recurrence-events',axis=1)
y=dum_cancer['Class_recurrence-events']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2021,stratify=y)
```

```python
nb=BernoulliNB(alpha=0.2,force_alpha=True)
nb.fit(X_train,y_train)   #Model building: Apriori probs calculation
```

```python
kfold=StratifiedKFold(n_splits=5,random_state=23,shuffle=True)
params={'alpha':np.linspace(0,5,20)}
```

```python
gcv=GridSearchCV(nb,param_grid=params,cv=kfold,scoring='neg_log_loss')
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```
