```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.model_selection import GridSearchCV,KFold
from sklearn.linear_model import ElasticNet
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
```

```python
chem=pd.read_csv("ChemicalProcess.csv")
X=chem.drop('Yield',axis=1)
y=chem['Yield']
```

```python
imp=SimpleImputer()
elastic=ElasticNet()
pipe=Pipeline([('impute',imp),('Elastic',elastic)])
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
pipe.get_params()
```

```python
params={'Elastic__l1_ratio':np.linspace(0.001,0.999,20),'Elastic__alpha':np.linspace(0.001,4,20)}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold)
gcv.fit(X,y)
```

```python
print("best parametprint("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)er :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

```python
med_imp=SimpleImputer(strategy='median')
elastic=ElasticNet()
pipe=Pipeline([('impute',med_imp),('Elastic',elastic)])
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```
