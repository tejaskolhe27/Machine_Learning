```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss 
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
from sklearn.ensemble import BaggingClassifier,VotingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
img=pd.read_csv("Image_Segmention.csv")
img.head()
```

```python
lbl=LabelEncoder()
img['Class']=lbl.fit_transform(img['Class'])
img.head()
```

```python
X=img.drop("Class",axis=1)
y=img['Class']
```

## LR

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=23,test_size=0.3,stratify=y)
```

```python
lr=LogisticRegression(multi_class='multinomial')
lr.fit(X_train,y_train)
```

```python
y_pred_proba=lr.predict_proba(X_test)
```

```python
lr_ll=log_loss(y_test,y_pred_proba)
print("Alone Logistic: ",lr_ll)
```

```python
bagg=BaggingClassifier(lr,n_estimators=15)
bagg.fit(X_train,y_train)
```

```python
y_pred_prob=bagg.predict_proba(X_test)
```

```python
bagg_ll=log_loss(y_test,y_pred_prob)
print("Bagging: ",bagg_ll)
```

# Using GCV

```python
lr=LogisticRegression()
svm=SVC(kernel='linear',probability=True,random_state=23)
dtc=DecisionTreeClassifier(random_state=23)
bagg=BaggingClassifier()
print(bagg.get_params())
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
params={'estimator':[lr,svm,dtc],
        'n_estimators':[10,15,30]}
gcv=GridSearchCV(bagg,param_grid=params,cv=kfold,scoring='neg_log_loss',verbose=3)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```
