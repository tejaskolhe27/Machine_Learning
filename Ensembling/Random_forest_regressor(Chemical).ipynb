```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss 
from sklearn.model_selection import train_test_split,GridSearchCV,KFold
from sklearn.ensemble import RandomForestRegressor,BaggingRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
chem=pd.read_csv("ChemicalProcess.csv")
chem.head()
```

```python
X=chem.drop("Yield",axis=1)
y=chem['Yield']
```

```python
imp=SimpleImputer()
imputed=imp.fit_transform(chem)
```

```python
chem_imp=pd.DataFrame(imputed,columns=chem.columns)
chem_imp
```

## 1. Simple Imputer(mean.median)

```python
imp_mean = SimpleImputer()
imputed_mean = imp_mean.fit_transform(chem)
imputed_mean
```

```python
imp_med = SimpleImputer(strategy="median")
imputed_median = imp_med.fit_transform(chem)
imputed_median
```

## 2. RForest Regressor

```python
imp=SimpleImputer()
rf=RandomForestRegressor(random_state=23)
pipe=Pipeline([('impute',imp),('RF',rf)])
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
pipe.get_params()
```

```python
params={'impute__strategy':['mean','median'],
        'RF__max_features':[2,3,4,5,8]}
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2',verbose=1)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

## 3. BaggingRegressor with ElasticNet

```python
elastic=ElasticNet()
bagg=BaggingRegressor(elastic,n_estimators=15)
```

```python
pipe=Pipeline([('impute',imp),('Bagging',bagg)])
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
pipe.get_params()
```

```python
params={'impute__strategy':['mean','median'],
        'Bagging__estimator__alpha':np.linspace(0.001,0.999,10),
        'Bagging__estimator__l1_ratio':np.linspace(0.001,4,10)}
gcv_bagg=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2',verbose=1)
gcv_bagg.fit(X,y)
```

```python
print("best parameter :",gcv_bagg.best_params_)
print("best score :",gcv_bagg.best_score_)
```
