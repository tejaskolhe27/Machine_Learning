```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("monthly-milk-production-pounds-p (1).csv")
```

```python
df.plot();
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
series = df['Milk']
result = seasonal_decompose(series,model='addictive',period =12)
result.plot()
plt.title("Addictive Decompostion")
plt.show()
```

```python
result = seasonal_decompose(series,model='multiplicative',period =12)
result.plot()
plt.title("Multiplicative Decompostion")
plt.show()
```
