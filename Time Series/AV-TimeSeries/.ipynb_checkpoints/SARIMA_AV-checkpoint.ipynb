#### **Bundus Bank Seasonal ARIMA**

```python
import pandas as pd
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
gdp = pd.read_csv("Train_SU63ISt.csv",index_col=1)
gdp.head()
```

```python
idx=pd.PeriodIndex(gdp.index,freq='Y')
gdp.index=idx
gdp.head()
```

```python
y = gdp['GDP per capita, PPP (constant 2011 international $)']
y_train = y[:-3]
y_test = y[-3:]
```

```python
from pmdarima.arima import auto_arima
```

```python
model = auto_arima(y_train, trace=True,
                   error_action='ignore', 
                   suppress_warnings=True,
                  seasonal= True,m=12)
```

```python
forecast = model.predict(n_periods=len(y_test))
forecast = pd.DataFrame(forecast,index = y_test.index,
                        columns=['Prediction'])
```

```python
from sktime.utils.plotting import plot_series
```

```python
#plot the predictions for validation set
plot_series(y_train,y_test,forecast,labels=['train','test','forecast'])
plt.show()
```

```python
rms = sqrt(mean_squared_error(y_test, forecast))
print('Test RMSE: %.3f' % rms)
```
