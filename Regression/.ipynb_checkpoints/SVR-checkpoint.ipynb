```python
from sklearn.svm import SVR
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,KFold,GridSearchCV
from sklearn.metrics import accuracy_score,log_loss
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
conc=pd.read_csv("Concrete_Data.csv")
conc.head()
```

```python
X=conc.drop('Strength',axis=1)
y=conc['Strength']
```

## RADIAL

```python
svr=SVR(kernel='rbf' )
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=23)
params={'C':np.linspace(0.001,5,10),'epsilon':np.linspace(0.001,5,10)}
gcv= GridSearchCV(svr,param_grid=params,cv=kfold,scoring='r2',verbose=2)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

```python
scaler=StandardScaler()
pipe=Pipeline([('SCL',scaler),('SVR',svr)])
pipe.get_params()
```

```python
params={'SVR__C':np.linspace(0.001,5,10),'SVR__epsilon':np.linspace(0.001,5,10)}
gcv= GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2',verbose=2)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

## LINEAR

```python
svr=SVR(kernel='linear')
```

```python
scaler=StandardScaler()
pipe=Pipeline([('SCL',scaler),('SVR',svr)])
pipe.get_params()
```

```python
params={'SVR__C':np.linspace(0.001,5,10),'SVR__epsilon':np.linspace(0.001,5,10)}
gcv= GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2',verbose=2,n_jobs=3)
gcv.fit(X,y)
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```

### Y Scaling (Linear)

```python
X=conc.drop('Strength',axis=1)
y=conc[['Strength']]
scaler_y=StandardScaler()
y_scl=scaler_y.fit_transform(y)
svr=SVR(kernel='linear')
```

```python
scaler=MinMaxScaler()
pipe=Pipeline([('SCL',scaler),('SVR',svr)])
pipe.get_params()
params={'SVR__C':np.linspace(0.001,5,10),'SVR__epsilon':np.linspace(0.001,5,10),
       'SCL':[StandardScaler(),MinMaxScaler()]}
gcv= GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='r2',verbose=2,n_jobs=3)
gcv.fit(X,y_scl[:,0])
```

```python
print("best parameter :",gcv.best_params_)
print("best score :",gcv.best_score_)
```
