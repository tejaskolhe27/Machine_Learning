```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LogisticRegression
from sklearn.metrics import accuracy_score,log_loss
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
cancer=pd.read_csv('BreastCancer.csv',index_col='Code')
```

```python
cancer.head()
```

```python
cancer=pd.get_dummies(dtype=int,data=cancer,drop_first=True)
cancer.head()
```

```python
X=cancer.drop('Class_Malignant',axis=1)
y=cancer['Class_Malignant']
```

```python
print(y.value_counts(normalize=True)*100)
print(y.value_counts())
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,test_size=0.3,random_state=23)
```

```python
print(y_train.value_counts(normalize=True)*100)
print(y_train.value_counts())
```

```python
print(y_test.value_counts(normalize=True)*100)
print(y_test.value_counts())
```

```python
lr=LogisticRegression()
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
res=cross_val_score(lr,X_train,y_train,cv=kfold)
print("Accuracy Score: ",res.mean()*100)
```

```python
res=cross_val_score(lr,X_train,y_train,cv=kfold,scoring='neg_log_loss')
print("Accuracy Score: ",res.mean()*100)
```

```python
res=cross_val_score(lr,X_train,y_train,cv=kfold,scoring='neg_mean_squared_error')
print("Accuracy Score: ",res.mean()*100)
```

```python
penalty={'penalty':['l1', 'l2', 'elasticnet', None],
         'solver':['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gcv=GridSearchCV(lr,param_grid=penalty,cv=kfold,scoring='neg_log_loss')
```

```python
gcv.fit(X,y)
```

```python
print(gcv.best_params_)
```

```python
print(gcv.best_score_)
```
