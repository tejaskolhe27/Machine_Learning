```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
housing  =  pd.read_csv("Housing.csv")
```

```python
housing.head()
```

```python
# label_encoder = LabelEncoder()
# housing.iloc['driveway'] = label_encoder.fit_transform(housing['driveway'])
```

```python
housing.head()
```

```python
housing = pd.get_dummies(housing,dtype=int)
```

```python
housing.head()
```

```python
X= housing.drop("price",axis =1)
y = housing['price']
```

```python
# Train,Test Split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
lr = LinearRegression()
dtr = DecisionTreeRegressor(random_state = 23)
```

```python
#Regression Model
model_lr = lr.fit(X_train,y_train)
y_pred = model_lr.predict(X_test)
```

```python
#Decision Tree Regressor Model
model_dtr = dtr.fit(X_train,y_train)
y_pred_dtr = model_dtr.predict(X_test)
```

```python
print("Linear regression",np.round(mean_squared_error(y_test,y_pred)))
print("Decision Tree Regressor",np.round(mean_squared_error(y_test,y_pred_dtr)))
```
