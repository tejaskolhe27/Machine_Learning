```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
housing =  pd.read_csv("Housing.csv")
```

```python
housing
```

```python
intervals = [(0,50000),(50000,100000),(100000,150000),(150000,200000)]
bins = pd.IntervalIndex.from_tuples(intervals)
housing['price_slab'] = pd.cut(housing['price'],bins)
```

```python
intervals = [(0,5000),(5000,10000),(10000,15000),(15000,20000)]
bins = pd.IntervalIndex.from_tuples(intervals)
housing['lotsize_slab'] = pd.cut(housing['lotsize'],bins)
```

```python
housing = housing.drop(['price','lotsize'],axis = 1)
```

```python
housing = housing.astype(object)
fd_df = pd.get_dummies(housing,prefix_sep='=')
```

```python
def gen_rules(min_sup,min_coef):
    itemsets=apriori(fd_df,min_support=min_sup,use_colnames=True)
    #and convert into rules
    rules=association_rules(itemsets,metric='confidence',min_threshold=min_coef)
    rules=rules[['antecedents','consequents','support','confidence','lift']]
    rules=rules[rules['lift']>1]
    return rules.sort_values(by='lift',ascending=False)
```

```python
import gradio as gr
demo=gr.Interface(gen_rules,
                 inputs=[gr.Slider(value=0.01,step=0.01,minimum=0.0001,maximum=1),
                        gr.Slider(value=0.01,step=0.01,minimum=0.0001,maximum=1)],
                 outputs='dataframe')

if __name__=="__main__":
    demo.launch()
```
