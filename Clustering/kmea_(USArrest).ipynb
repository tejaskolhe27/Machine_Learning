```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
milk=pd.read_csv("milk.csv",index_col=0)
milk.shape
```

```python
scaler=StandardScaler()
milk_scaled=scaler.fit_transform(milk)
```

```python
clust=[2,3,4,5,6,7,8]
scores=[]
```

```python
for i in clust:
    km=KMeans(n_clusters=i, random_state=23)
    km.fit(milk_scaled)
    scores.append(km.inertia_)
```

```python
plt.scatter(clust,scores, c='red')
plt.plot(clust,scores, c='blue')
plt.title('Screen plot')
plt.xlabel('Np. of clusters')
plt.ylabel('WSS')
plt.show()
```

1. According to screen plot, point no. 4 is the elbow point.
2. Which means that 4 number of the cluster will be the optimal number of clusering.
