```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import r2_score,accuracy_score
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
from sklearn.ensemble import BaggingRegressor,RandomForestClassifier,StackingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cancer=pd.read_csv("BreastCancer.csv",index_col=0)
X=cancer.drop('Class',axis=1)
y=cancer['Class']
```

```python
tsne=TSNE(random_state=23,perplexity=50).set_output(transform='pandas')
```

```python
X_tsne=tsne.fit_transform(X)
```

```python
tsne_pd=pd.concat([X_tsne,y],axis=1)
```

```python
sns.scatterplot(data=tsne_pd,x='tsne0',y='tsne1',hue='Class')
plt.title("t-SNE")
plt.show()  # t-SNE has better classification than PCA
```

```python
svm=SVC(kernel='linear',random_state=23)
```

```python
pipe=Pipeline([('TSNE',tsne),('SVM',svm)])
pipe.get_params()
```

```python
kfold=StratifiedKFold(n_splits=5,shuffle=True,random_state=23)
```

```python
params={'TSNE__perplexity':[5,15,30,35]}
```

```python
gcv=GridSearchCV(pipe,param_grid=params,cv=kfold,scoring='accuracy',verbose=3)
gcv.fit(X,y)
```
